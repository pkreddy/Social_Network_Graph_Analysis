## Loading and Manipulating graphs using Networkx

```python
import networkx as nx
import pandas as pd
import numpy as np
%matplotlib notebook
```

```python
#Adding Nodes
G1 = nx.Graph()
G1.add_edge(0,1)
nx.draw_networkx(G1)
```

```python
#Adding Edge Attributes
G1 = nx.Graph()
G1.add_edge('A','B',weight=6,relation='family')
G1.add_edge('B','C',weight=3,relation='friend')
nx.draw_networkx(G1)
print(G1.edges())
print(G1.edges(data=True))
print(G1.edges(data='weight'))
```

```python
glist = [(1,2),(2,3),(3,4),(4,1),(1,3),(2,4)]
G1 = nx.Graph()
G1.add_edges_from(glist)
nx.draw_circular(G1)
```

```python
G1 = nx.Graph()
# add node/edge pairs
G1.add_edges_from([(0, 1),
                   (0, 2),
                   (0, 3),
                   (0, 5),
                   (1, 3),
                   (1, 6),
                   (3, 4),
                   (4, 5),
                   (4, 7),
                   (5, 8),
                   (8, 9)])
nx.draw_networkx(G1)
```
